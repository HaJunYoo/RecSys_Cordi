## 1. 코디 결측치 채우기

> - 1차 브랜드로 그룹핑 
> - 2차 인덱스 main_category + sub_category
> - 코디 forwardfiill 같은 걸로 결측값 채우기

- 이렇게 해결 안될 것 같음
> coordi 값 있는 것들 dictinary 형태로 만들고 
없는 값 채우는 형식으로 해보자 


```python
#corrdi_값 있는 것들 따로 저장

coordi_notna = new_musinsa[new_musinsa["coordi"].notnull()][["brand", "main_category","sub_category","coordi"]].reset_index()
```

```python
#key 값: brand+main_category+sub_category / value 값: coordi로 dictionary 생성
#brand, category 같아도 코디 다른 경우 있음 -> 어떤 것으로 대체? 일단 합칩합으로 실행  

import ast

coordi_notna_dic = {}
for i in range(coordi_notna.shape[0]):
    brand = coordi_notna.loc[i,"brand"]
    main = coordi_notna.loc[i,"main_category"]
    sub = coordi_notna.loc[i,"sub_category"]
    key = brand + main + sub
    value = ast.literal_eval(coordi_notna.loc[i,"coordi"])
    temp = coordi_notna_dic.get(key)
    if temp == None:
        coordi_notna_dic[key] = value
    else:
        coordi_notna_dic[key] = list(set(temp) | set(value))
```

```python
coordi_notna_dic
```

```python
for i in range(new_musinsa.shape[0]):
    brand = new_musinsa.loc[i,"brand"]
    main = new_musinsa.loc[i,"main_category"]
    sub = new_musinsa.loc[i,"sub_category"]
    key = brand + main + sub
    if pd.isna(new_musinsa.loc[i, "coordi"]):
        try:
            new_musinsa.loc[i, "coordi"] = coordi_notna_dic[key]
        except:
            pass
```

```python
#tags
import ast

a = []
num_na = 0

for i in musinsa_21["tags"]:
    li = ast.literal_eval(i)
    for j in li:
        a.append(j)
    if len(li) == 0:
        num_na += 1
```

```python
#coordi

a = []
num_na = 0

for i in musinsa_21["coordi"]:
    try:
        li = ast.literal_eval(i)
        for j in li:
            a.append(j)
    except:
        num_na += 1
```

```python
print(len(a))
print(len(list(set(a))))
print(num_na)
```

###### tags는 너무 많으니까 보류 
###### coordi는 고유값 12개 TF-IDF 해보자 

```python
#tags는 너무 많으니까 보류 
#coordi는 고유값 12개 TF-IDF 해보자 
```

```python
total_count = len(musinsa_21.index)
total_coordi = list(set(a))
```

```python
coordi_count = dict.fromkeys(total_coordi)

for each_coordi in musinsa_21["coordi"]:
    try:
        for coordi in ast.literal_eval(each_coordi):
            if coordi_count[coordi] == None:
                coordi_count[coordi] = 1
            else:
                coordi_count[coordi] += 1
    except:
        pass
```

```python
coordi_count
```

```python
## IDF 해당하는 부분 -> 장르별 가중치 생성하겠다. 

for coordi in coordi_count:
    coordi_count[coordi] = np.log10(total_count/coordi_count[coordi]) 
  
coordi_count 
```

```python
from tqdm import tqdm

coordi_representation = pd.DataFrame(columns=sorted(total_coordi), index=musinsa_21.index)
for index, each_row in tqdm(musinsa_21.iterrows()):
    try:
        dict_temp = {i: coordi_count[i] for i in ast.literal_eval(each_row["coordi"])}
    except:
        dict_temp = {i: np.NaN}
    row_to_add = pd.DataFrame(dict_temp, index=[index])
    coordi_representation.update(row_to_add)

coordi_representation
```

```python
coordi_representation.fillna(0, inplace=True)
coordi_representation
```

```python
#유사도 평가 - cosine similarity

from sklearn.metrics.pairwise import cosine_similarity

def cos_sim_matrix(a, b):
    cos_sim = cosine_similarity(a, b)
    result_df = pd.DataFrame(data=cos_sim, index=[a.index])

    return result_df
```

```python
print(coordi_representation.head())
```

```python
cs_df = cos_sim_matrix(coordi_representation, coordi_representation)
cs_df.head()
```
